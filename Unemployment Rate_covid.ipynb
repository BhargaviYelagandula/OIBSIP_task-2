```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
data=pd.read_csv("C:/Users/Bhargavi/Desktop/Unemployment_Rate_upto_11_2020.csv")
data.columns=["State","Date","Frequency","Estimated Unemployment","Estimated Employed","Estimated Labour participation","Region","Longitude","Latitude"]
print(data.head())
```

```python
print(data.isnull().sum())
```

```python
data.Region.nunique()
```

```python
plt.title("Indian Unemployment")
sns.histplot(x="Estimated Employed", hue="Region", data=data)
plt.show()
```

```python
plt.figure(figsize=(12, 10))
plt.title("Indian Unemployment")
sns.histplot(x="Estimated Unemployment", hue="Region", data=data)
plt.show()
```

```python
plt.figure(figsize=(12, 10))
plt.title("Indian Unemployment Region Wise")
sns.barplot(x="Region",y="Estimated Unemployment", data=data)
plt.show()
```

```python
unemploment = data[["State", "Region", "Estimated Unemployment"]]
figure = px.sunburst(unemploment, path=["Region", "State"], 
                     values="Estimated Unemployment", 
                     width=700, height=700, color_continuous_scale="RdY1Gn", 
                     title="Unemployment Rate in India")
figure.show()
```
